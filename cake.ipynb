## Ingredients
The first step is to gather all the ingredients. For a basic cake, you'll need the following:
- 2 cups all-purpose flour
- 2 teaspoons baking powder
- 1/2 teaspoon salt
- 1/2 cup unsalted butter, softened
- 1 cup white granulated sugar
- 2 large eggs
- 1 cup whole milk
- 1 teaspoon vanilla extract

## Directions
1. Preheat your oven to 350°F (175°C) and prepare a 9-inch cake pan by greasing it with butter or cooking spray.
2. In a medium-sized bowl, whisk together the flour, baking powder, and salt.
3. In a separate large bowl, cream together the softened butter and sugar until light and fluffy.
4. Add the eggs, one at a time, to the butter mixture, beating well after each addition.
5. Gradually add the flour mixture to the butter mixture, alternating with the milk, beginning and ending with the flour mixture. Mix until just combined.
6. Stir in the vanilla extract.
7. Pour the batter into the prepared cake pan and smooth the surface with a spatula.
8. Bake the cake for 25-30 minutes, or until a toothpick inserted into the center comes out clean.
9. Let the cake cool in the pan for 10 minutes before transferring it to a wire rack to cool completely.
10. Once the cake is cool, you can frost it with your favorite frosting or serve it plain.

## Tips and Tricks
- Make sure all of your ingredients are at room temperature before you begin baking. This will ensure that they combine evenly and create a smooth batter.
- Use a kitchen scale to measure your ingredients accurately. Baking is a science, and precise measurements are crucial for success.
- Don't overmix the batter. Once the flour is added, mix it until just combined. Overmixing can lead to a tough, dense cake.
- Keep an eye on your cake as it bakes. Every oven is different, so the baking time may vary. Check the cake after 20 minutes and then every 5 minutes after that.
- Let the cake cool completely before frosting it. If the cake is still warm, the frosting will melt and slide off.
- Have fun and experiment with different flavors and decorations. Baking is a creative outlet, so don't be afraid to try new things!
