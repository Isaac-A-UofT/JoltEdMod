# Loops in Python

In programming, loops are used to execute a set of instructions repeatedly until a particular condition is met. Python has two types of loops - `for` and `while` loops.

## For Loops

A `for` loop is used to iterate over a sequence of elements. The sequence can be a list, tuple, set, or a string. Let's look at an example:

```python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

In this example, we have a list of fruits and we use a `for` loop to iterate over each element of the list and print it. The output will be:

```
apple
banana
cherry
```

We use the `in` keyword to define the sequence we want to iterate over. The variable `fruit` is used to hold the current element of the sequence in each iteration of the loop.

## While Loops

A `while` loop is used to execute a set of instructions repeatedly as long as a particular condition is true. Let's look at an example:

```python
i = 1
while i <= 5:
    print(i)
    i = i + 1
```

In this example, we initialize a variable `i` to 1 and use a `while` loop to print the value of `i` and increment it by 1 in each iteration, until `i` is no longer less than or equal to 5. The output will be:

```
1
2
3
4
5
```

We use the `while` keyword to define the condition that needs to be true for the loop to continue executing. In each iteration of the loop, we need to update the value of the variable that defines the condition, otherwise the loop will execute indefinitely.

## Conclusion

Loops are an essential part of programming, allowing us to execute a set of instructions repeatedly until a certain condition is met. `for` loops are used to iterate over a sequence of elements, while `while` loops are used to execute a set of instructions repeatedly as long as a particular condition is true.
