### What are Conditional Statements?

Conditional statements are used to execute certain blocks of code based on whether a certain condition is true or false. They are also known as if-else statements.

### Syntax

The syntax for conditional statements in most programming languages is as follows:

```python
if condition:
    statement1
else:
    statement2
```

In the above syntax, `condition` is the expression that is evaluated to either `True` or `False`. If `condition` is `True`, then `statement1` is executed. Otherwise, `statement2` is executed.

### Example

Let's take an example to understand this better. Suppose we want to write a program that checks whether a given number is even or odd. Here's how we would do it using conditional statements:

```python
# take input from the user
num = int(input("Enter a number: "))

# check if the number is even or odd
if num % 2 == 0:
    print(num, "is even")
else:
    print(num, "is odd")
```

In the above code, we take input from the user using the `input()` function and convert it to an integer using the `int()` function. We then check if the number is even or odd using the `%` operator. If the remainder when the number is divided by 2 is 0, then the number is even and we print `"is even"`. Otherwise, we print `"is odd"`.

### Conclusion

Conditional statements are a very important concept in programming and are used extensively. They allow us to write programs that can make decisions based on certain conditions, making our programs more dynamic and flexible.
