# Installing Python

Python is an open-source programming language that is widely used in the world of computer science. It is easy to learn and has a large community of developers who contribute to its development. In this tutorial, we will learn how to install Python on your computer.

## Downloading Python

The first step to installing Python is to download the installer from the official website (https://www.python.org/downloads/).

Here's how to download and install Python on your computer:

1. Go to the official Python website (https://www.python.org/downloads/) and download the latest version of Python.

2. Once downloaded, run the installer and follow the instructions.

3. When prompted, select the option to add Python to your PATH environment variable. This will make it easier to run Python from the command line.

## Checking the Installation

To check if Python has been installed correctly on your computer, open a terminal window and type the following command:

```python
python --version
```

This command should display the version of Python that you have installed on your computer.

## Running Python

Once Python has been installed, you can start using it to write and run your own programs.

To run a Python program, create a new file with a .py extension and write your code using a text editor. Then, save the file and navigate to the directory where the file is saved using the command line.

To run the program, type the following command:

```python
python filename.py
```

Replace "filename.py" with the name of the file that you saved earlier.

## Conclusion

That's it! You have now learned how to download and install Python on your computer. Now, you can start exploring the world of Python programming and develop your own applications.
