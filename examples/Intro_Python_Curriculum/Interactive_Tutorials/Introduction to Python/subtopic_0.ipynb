## Example Code

Let's take a look at a simple Python program that demonstrates some of the basic features of the language. This program will ask the user for their name, and then greet them with a personalized message.

```python
# Ask the user for their name
name = input("What is your name? ")

# Greet the user with a personalized message
print("Hello, " + name + "! Welcome to Python.")
```

In this code, we use the `input()` function to ask the user for their name. The `input()` function allows us to get input from the user, and returns it as a string. We then store the user's name in a variable called `name`.

Next, we use the `print()` function to display a personalized message to the user. The `print()` function allows us to output text to the console. We use the `+` operator to concatenate the user's name with the rest of the message.

This is a very simple example, but it demonstrates some of the basic features of Python. We can use variables to store data, and use functions to manipulate that data. We can also get input from the user and output information to the console.
