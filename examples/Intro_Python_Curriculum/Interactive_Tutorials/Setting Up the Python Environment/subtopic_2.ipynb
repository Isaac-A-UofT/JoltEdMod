# Using the command line

Welcome, first year computer science students!

Today, we're going to talk about using the command line. The command line is a way to interact with your computer using text commands.

## Why use the command line?

You might be wondering, why would I want to use the command line when I can use a graphical user interface (GUI) like Windows or macOS? Well, there are a few reasons:

- **Speed:** Once you get the hang of it, using the command line can be much faster than using a GUI. You can navigate to different directories, create and delete files, and perform other tasks with just a few keystrokes.

- **Power:** The command line gives you access to a wide range of tools and utilities that you might not have access to otherwise. For example, you can use command line tools to automate tasks, manipulate large datasets, and more.

- **Flexibility:** The command line is highly customizable, and you can create your own scripts and commands to perform tasks that are specific to your needs.

## Basic commands

Let's start by looking at some basic commands that you can use in the command line. Open up a terminal or command prompt to follow along.

### `cd`

The `cd` command is used to change directories. For example, if you want to navigate to your home directory, you can type:

```
cd ~
```

The `~` symbol is a shortcut for your home directory. You should now be in your home directory.

### `ls`

The `ls` command is used to list the contents of a directory. For example, if you want to list the contents of your home directory, you can type:

```
ls
```

You should see a list of files and directories in your home directory.

### `mkdir`

The `mkdir` command is used to create a new directory. For example, if you want to create a new directory called "my_folder", you can type:

```
mkdir my_folder
```

You should now have a new directory called "my_folder" in your current directory.

### `touch`

The `touch` command is used to create a new file. For example, if you want to create a new file called "my_file.txt", you can type:

```
touch my_file.txt
```

You should now have a new file called "my_file.txt" in your current directory.

### `rm`

The `rm` command is used to delete files. For example, if you want to delete the file "my_file.txt", you can type:

```
rm my_file.txt
```

The file should now be deleted.

## Conclusion

That's it for our brief introduction to using the command line. These are just a few basic commands, but there are many more that you can use to perform a wide range of tasks.
