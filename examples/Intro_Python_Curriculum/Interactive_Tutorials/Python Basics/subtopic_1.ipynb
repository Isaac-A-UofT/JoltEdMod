# Variables and Data Types

Welcome to the first lecture of Computer Science 101! Today, we will be discussing one of the most fundamental concepts in programming: variables and data types.

## Variables

A variable is a container that holds a value. Think of it as a labeled box that you can put things into and take things out of. In programming, we use variables to store and manipulate data.

Let's take a look at an example:

```python
# This is a variable named "my_variable" that holds the value "Hello, World!"
my_variable = "Hello, World!"

# We can print the value of the variable to the console
print(my_variable)
```

In this example, we create a variable named `my_variable` and assign it the value `"Hello, World!"`. We then print the value of the variable to the console.

## Data Types

A data type is a classification of data that determines the type of operations that can be performed on it. There are many different data types in programming, but some of the most common ones include strings, integers, and floats.

### Strings

A string is a sequence of characters. It is used to represent text in programming. In Python, strings are surrounded by either single or double quotes.

```python
# This is a string variable named "my_string" that holds the value "Hello, World!"
my_string = "Hello, World!"

# We can print the value of the variable to the console
print(my_string)
```

In this example, we create a string variable named `my_string` and assign it the value `"Hello, World!"`. We then print the value of the variable to the console.

### Integers

An integer is a whole number. It is used to represent numbers in programming.

```python
# This is an integer variable named "my_integer" that holds the value 42
my_integer = 42

# We can print the value of the variable to the console
print(my_integer)
```

In this example, we create an integer variable named `my_integer` and assign it the value `42`. We then print the value of the variable to the console.

### Floats

A float is a decimal number. It is used to represent numbers with a fractional component in programming.

```python
# This is a float variable named "my_float" that holds the value 3.14
my_float = 3.14

# We can print the value of the variable to the console
print(my_float)
```

In this example, we create a float variable named `my_float` and assign it the value `3.14`. We then print the value of the variable to the console.
