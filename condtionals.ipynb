# Chapter 2: Conditional Statements in Python

In this chapter, we will learn about conditional statements in Python. 

## Introduction

Conditional statements are used in programming to make decisions based on certain conditions. In Python, we use the `if` statement to create a conditional statement. 

## Example

Let's consider an example where we want to print a message if a number is greater than 5. 

```python
number = 6

if number > 5:
    print("The number is greater than 5")
```

In the above code, we assign the value 6 to the variable `number`. We then use an `if` statement to check if `number` is greater than 5. If the condition is true, i.e., if `number` is greater than 5, we print the message "The number is greater than 5".

## Explanation

In the above example, we use a comparison operator `>` to check if `number` is greater than 5. The `if` statement then checks if this condition is true. If it is true, the code inside the `if` statement is executed. 

If the condition is false, i.e., if `number` is not greater than 5, the code inside the `if` statement is not executed. 

We can also use the `else` statement to execute code when the condition in the `if` statement is false. For example:

```python
number = 2

if number > 5:
    print("The number is greater than 5")
else:
    print("The number is not greater than 5")
```

In the above code, since `number` is not greater than 5, the condition in the `if` statement is false. Therefore, the code inside the `else` statement is executed and we print the message "The number is not greater than 5". 

## Conclusion

Conditional statements are an important part of programming and are used to make decisions based on certain conditions. In Python, we use the `if` statement to create a conditional statement. We can also use the `else` statement to execute code when the condition in the `if` statement is false.
